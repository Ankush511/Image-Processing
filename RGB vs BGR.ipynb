```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

## 1. Loading an Image using OpenCV

```python
img = cv2.imread('transition.jpg')
plt.imshow(img)
```

* BGR <-> RGB
* RGB <-> BGR

## 2. Converting BGR to RGB

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)
```

```python
cv2.imwrite('BGRimg.jpg', img)
```

```python
img.shape
```

## 3. Frame Extraction

```python
plt.imshow(img[:, :, :])
```

```python
r = img[:, :, 0] # Red
g = img[:, :, 1] # Green
b = img[:, :, 2] # Blue
```

```python
plt.imshow(r)
```

```python
plt.imshow(g)
```

```python
plt.imshow(b)
```

```python
cv2.imwrite('Red.jpg',r)
cv2.imwrite('Green.jpg',g)
cv2.imwrite('Blue.jpg',b)
```

## 4. Loading and Displaying Image using OpenCV Window

```python
img = cv2.imread('transition.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2.imshow('Image', img)
cv2.waitKey(0)
```
